```python
exp = df.loc[(df['experiment_id'] == 'b382c6') & (df['metric_id'] == 1) ]
print(exp.shape)
exp.head(10)
```

```python
row = exp.loc[16581]
mean_c = row['mean_c']      
mean_t = row['mean_t']       
var_c = row['variance_c']    
var_t = row['variance_t']   
n_c = row['count_c']        
n_t = row['count_t'] 
```

```python
se = np.sqrt((var_c / n_c) + (var_t / n_t))
print(f"Standard Error: {se:.6f}")
```

```python
## 6. Calculate Z-Score

z = (mean_t - mean_c) / se
print(f"Z-score: {z:.4f}")
```

```python
## 7. Calculate P-value

p_value = 2 * norm.sf(abs(z))
print(f"P-value: {p_value:.6f}")
```
